```python

import pandas as pd
import numpy as np
```

```python
from sklearn.impute import SimpleImputer
```

```python
lst1=['France',44.0,72000.0,'no']
lst2=['Spain',27.0,48000.0,"Yes"]
lst3=['india',30.00,'NaN',]
```

```python
lst4=["Germany",'NaN',57000.0,"Yes"]
lst5=["usa",50.0,83000.0,"yes"]
```

```python
df=pd.DataFrame([lst1,lst2,lst3,lst4,lst5],columns=['country','age','salary','purchased'])
```

```python
df
```

```python
df.dropna()
```

```python
#introduction to imputers for removing rows with null values
#strategy has mode mean median most freqquent,etc
```

```python
imputer=SimpleImputer(missing_values=np.nan,strategy='mean')
```

```python
np.array(df["salary"])
```

```python
#above error is shown as imputer needs a 2d array always so 1d array needs to nbe sliced
```

```python
#now to get more accurate analysis we use most_frequent instead of mean
imputer = SimpleImputer(missing_values = np.nan, strategy = 'most_frequent')

imputer.fit(df.iloc[:,:].values)

df.iloc[:,:] = imputer.transform(df.iloc[:,:].values)
```
