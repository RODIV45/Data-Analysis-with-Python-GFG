```python
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer
```

```python
from google.colab import drive
drive.mount("/content/drive")
```

```python
path="/content/drive/MyDrive/airports.csv"
df=pd.read_csv(path)
```

```python
df
```

```python
df.head()
```

```python
df.isnull().sum()
#shows how many null rows under each column
```

```python
df.iloc[:,6:7].values
```

```python
impute=SimpleImputer(missing_values=np.nan,strategy='mean')
impute.fit(df.iloc[:,6:7].values)
```

```python
df.iloc[:,6:7]=impute.transform(df.iloc[:,6:7].values)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
#see clearly above that elevation has np null values
```

```python
impute=SimpleImputer(missing_values=np.nan,strategy='most_frequent')
impute.fit(df.iloc[:,:].values)
```

```python
df.iloc[:,:]=impute.transform(df.iloc[:,:].values)
```
